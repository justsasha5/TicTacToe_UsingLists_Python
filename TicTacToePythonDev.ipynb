## Tic Tac Toe  Using Lists


```python
line1 = ["⬜️","️⬜️","️⬜️"]
line2 = ["⬜️","⬜️","️⬜️"]
line3 = ["⬜️️","⬜️️","⬜️️"]
map = [line1, line2, line3]

# input should be like X12 (where 1 is row and 2 is column)
print("Input X or O and number for row and column respectively where you want to put it:")
print("P.S. Your input should look like ex: X12 (where 1 is row and 2 is column)")


counter = 0
var=True

while counter<9 and var == True:
    position = input() 
#Below line to break input in list by element
    INlist = list(position)
    
#Below lines preventative measures for incorrect input
    try:
        letter = INlist[0].lower()
        row_index = int(INlist[1]) - 1
        column_index = int(INlist[2]) - 1
    except:
        print("Incorrect Input. please input letter (X or O) and 2 numbers for row and column, numbers 1, 2 or 3")
        quit()
    if letter != "x" and letter  != "o":
        print("Incorrect Input. please input letter (X or O)")
        var=False
    elif int(INlist[1])>3 or int(INlist[2])>3 or 1>int(INlist[1]) or 1>int(INlist[2]):
        print("Incorrect Input. please input 2 numbers for row and column, numbers 1, 2 or 3")
        var=False
    elif map[row_index][column_index] =="o" or map[row_index][column_index] =="x":
        print(f"Player {letter} Choose another row and column as value already filled.")
    else:
    #Below line puts X or O in the correct list on our grid
        map[row_index][column_index] = INlist[0]
        counter+=1
        print(f"{line1}\n{line2}\n{line3}")

    #Cover Diagonals *2
        if map[0][0] == "x" and map[1][1] == "x" and map[2][2] == "x":
            print("player 1 (x) won!")
            break
        elif map[0][2] == "x" and map[1][1] == "x" and map[2][0] == "x":
            print("player 1 (x) won!")
            break     
    #Cover Rows *3
        elif map[0][0] == "x" and map[0][1] == "x" and map[0][2] == "x":
            print("player 1 (x) won!") 
            break        
        elif map[1][0] == "x" and map[1][1] == "x" and map[1][2] == "x":
            print("player 1 (x) won!")
            break
        elif map[2][0] == "x" and map[2][1] == "x" and map[2][2] == "x":
            print("player 1 (x) won!")      
            break
    #Cover Columns *3        
        elif map[0][0] == "x" and map[1][0] == "x" and map[2][0] == "x":
            print("player 1 (x) won!")
            break
        elif map[0][1] == "x" and map[1][1] == "x" and map[2][1] == "x":
            print("player 1 (x) won!")
            break
        elif map[0][2] == "x" and map[1][2] == "x" and map[2][2] == "x":
            print("player 1 (x) won!")
            break

    #Now same for o        
        elif map[0][0] == "o" and map[1][1] == "o" and map[2][2] == "o":
            print("player 2 (o) won!")
            break
        elif map[0][2] == "o" and map[1][1] == "o" and map[2][0] == "o":
            print("player 2 (o) won!")    
            break
        elif map[0][0] == "o" and map[0][1] == "o" and map[0][2] == "o":
            print("player 2 (o) won!") 
            break
        elif map[1][0] == "o" and map[1][1] == "o" and map[1][2] == "o":
            print("player 2 (o) won!")
            break
        elif map[2][0] == "o" and map[2][1] == "o" and map[2][2] == "o":
            print("player 2 (o) won!")      
            break
    #Cover Columns *3        
        elif map[0][0] == "o" and map[1][0] == "o" and map[2][0] == "o":
            print("player 2 (o) won!")
            break
        elif map[0][1] == "o" and map[1][1] == "o" and map[2][1] == "o":
            print("player 2 (o) won!")
            break
        elif map[0][2] == "o" and map[1][2] == "o" and map[2][2] == "o":
            print("player 2 (o) won!")
            break
print(f"{line1}\n{line2}\n{line3}")
```
